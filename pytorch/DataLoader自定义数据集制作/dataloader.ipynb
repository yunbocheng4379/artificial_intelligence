```python
def load_annotations(ann_file):
    data_infos = {}
    with open(ann_file) as f:
        samples = [x.strip().split(' ') for x in f.readlines()]
        for filename, gt_label in samples:
            data_infos[filename] = np.array(gt_label, dtype=np.int64)
    return data_infos
```

#### 任务4：把上面那几个事得写在一起

- 1.注意要使用from torch.utils.data import Dataset, DataLoader
- 2.类名定义class FlowerDataset(Dataset)，其中FlowerDataset可以改成自己的名字
- 3.def __init__(self, root_dir, ann_file, transform=None):咱们要根据自己任务重写
- 4.def __getitem__(self, idx):根据自己任务，返回图像数据和标签数据

```python
from torch.utils.data import Dataset, DataLoader
class FlowerDataset(Dataset):
    # 第一步：构建数据集和标签的两个List集合
    def __init__(self, root_dir, ann_file, transform=None):
        # 存储图片和图片标签的文件
        self.ann_file = ann_file
        # 图片根路径
        self.root_dir = root_dir
        # 将图片和图片标签存储为一个字典
        self.img_label = self.load_annotations()
        # 将图片的路径存储到一个list集合中
        self.img = [os.path.join(self.root_dir,img) for img in list(self.img_label.keys())]
        # 将图片的标签存储到一个list集合中
        self.label = [label for label in list(self.img_label.values())]
        # 声明传递进行的transformer框架
        self.transform = transform
 
    def __len__(self):
        return len(self.img)

    # idx是一个随机数，代表一个下标值，从图片和标签集合中获取指定下标对应的值
    def __getitem__(self, idx):
        # 第二步：将数据集和标签返回（tensor）
        image = Image.open(self.img[idx])
        label = self.label[idx]
        if self.transform:
            # 使用transformer给图片做预处理操作
            # 预处理之后图片会转换为 tensor 格式
            image = self.transform(image)
        # 标签不需要进行预处理操作，将标签转换为tensor格式
        label = torch.from_numpy(np.array(label))
        return image, label
    def load_annotations(self):
        data_infos = {}
        with open(self.ann_file) as f:
            samples = [x.strip().split(' ') for x in f.readlines()]
            for filename, gt_label in samples:
                data_infos[filename] = np.array(gt_label, dtype=np.int64)
        return data_infos
```

#### 任务5：数据预处理(transform)

- 1.预处理的事都在上面的__getitem__中完成，需要对图像和标签咋咋地的，要整啥事，都在上面整
- 2.返回的数据和标签就是建模时模型的输入和损失函数中标签的输入，一定整明白自己模型要啥
- 3.预处理这个事是你定的，不同的数据需要的方法也不一样，下面给出的是比较通用的方法

```python
data_transforms = {
    'train': 
        transforms.Compose([
        transforms.Resize(64),
        transforms.RandomRotation(45),#随机旋转，-45到45度之间随机选
        transforms.CenterCrop(64),#从中心开始裁剪
        transforms.RandomHorizontalFlip(p=0.5),#随机水平翻转 选择一个概率概率
        transforms.RandomVerticalFlip(p=0.5),#随机垂直翻转
        transforms.ColorJitter(brightness=0.2, contrast=0.1, saturation=0.1, hue=0.1),#参数1为亮度，参数2为对比度，参数3为饱和度，参数4为色相
        transforms.RandomGrayscale(p=0.025),#概率转换成灰度率，3通道就是R=G=B
        transforms.ToTensor(),
        # 标准化操作：使用大数据集计算出的标准差来应用到本数据集（因为本数据集较小，计算出来的标准差不准确），
        # 均值 [0.485, 0.456, 0.406]：分别对应 RGB 三个通道 的像素平均值。（除法）
        # 标准差 [0.229, 0.224, 0.225]：分别对应 RGB 三个通道 的像素标准差。（减法）
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])#均值，标准差
    ]),
    'valid': 
        transforms.Compose([
        transforms.Resize(64),
        transforms.CenterCrop(64),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}
```

#### 任务7：用之前先试试，整个数据和标签对应下，看看对不对

- 1.别着急往模型里传，对不对都不知道呢
- 2.用这个方法：iter(train_loader).next()来试试，得到的数据和标签是啥
- 3.看不出来就把图画出来，标签打印出来，确保自己整的数据集没啥问题

```python
# iter(train_loader)：将 DataLoader 对象转换为一个 迭代器（Iterator），用于按顺序遍历数据。
# next(...)：从迭代器中取出 下一个批次的数据.
# 之所以每次都可以获取到不同的随机数据，并不是因为以上两个函数的原因，而是因为使用DataLoader()函数加载数据时，
# 设置了shuffle=True参数，每次执行都会随机打乱数据分布，导致获取到的数据都是随机的
# 此时获取的并不是一张图片，而是一批（batch）图片
image, label = next(iter(train_loader))
# image[0]：获取一个batch中的第一张图片
sample = image[0].squeeze()
# 将张量转换为 NumPy 数组并调整通道顺序
# plt在绘制图像时，需要按照【宽度 × 高度 × 通道】格式进行绘制，需要从 C×H×W 转为 H×W×C
sample = sample.permute((1, 2, 0)).numpy()
# 反归一化：在数据预处理时，使用了transforms.Normalize
# 需要将之前均值(乘法)和标准差补回来(加法)
sample *= [0.229, 0.224, 0.225]
sample += [0.485, 0.456, 0.406]
# imshow()：将数据（数组或张量）渲染为图像，但不会直接显示图像
plt.imshow(sample)
# show()：渲染并显示所有已创建的图像/图形（如 plt.imshow()、plt.plot() 等）
plt.show()
print('Label is: {}'.format(label[0].numpy()))
```

```python
# 优化器设置
optimizer_ft = optim.Adam(model_ft.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)#学习率每7个epoch衰减成原来的1/10
# 衰减函数设置
criterion = nn.CrossEntropyLoss()
```

```python
# 初始化模型
def train_model(model, dataloaders, criterion, optimizer, num_epochs=25, is_inception=False, filename='best.pth'):
    # 模型预测起始时间
    since = time.time()
    # 记录训练最好一次的epoch
    best_acc = 0
    # 判断使用哪种方式进行计算
    model.to(device)

    val_acc_history = []
    train_acc_history = []
    train_losses = []
    valid_losses = []
    # 记录学习率
    LRs = [optimizer.param_groups[0]['lr']]
    # 记录训练最好一次的模型
    best_model_wts = copy.deepcopy(model.state_dict())

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        # 训练和验证
        for phase in ['train', 'valid']:
            if phase == 'train':
                model.train()  # 训练
            else:
                model.eval()   # 验证

            # 统计损失率
            running_loss = 0.0
            # 统计预测成功个数
            running_corrects = 0

            # 把数据都取个遍
            for inputs, labels in dataloaders[phase]:
                # 将数据集和标签存储到CPU/GPU中
                inputs = inputs.to(device)
                labels = labels.to(device)

                # 梯度清零
                optimizer.zero_grad()
                # 只有训练的时候计算和更新梯度
                with torch.set_grad_enabled(phase == 'train'):
                    # 调用模型预测结果
                    outputs = model(inputs)
                    # 调用损失函数计算损失
                    loss = criterion(outputs, labels)
                    # 获取输出结果中预测比重最大的作为最终预测结果
                    _, preds = torch.max(outputs, 1)
                    #print(loss)

                    # 训练阶段更新权重
                    if phase == 'train':
                        # 反向传播
                        loss.backward()
                        # 梯度更新（权重、参数更新）
                        optimizer.step()

                # 统计每批batch数据集的损失值
                running_loss += loss.item() * inputs.size(0)
                # 统计每批batch数据集预测成功的个数
                running_corrects += torch.sum(preds == labels.data)

            # 计算本次epoch的平均损失率
            epoch_loss = running_loss / len(dataloaders[phase].dataset)
            # 计算本次epoch的预测成功率
            epoch_acc = running_corrects.double() / len(dataloaders[phase].dataset)
            
            # 计算本次epoch预测时长
            time_elapsed = time.time() - since
            print('Time elapsed {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
            print('{} Loss: {:.4f} Acc: {:.4f}'.format(phase, epoch_loss, epoch_acc))

            # 计算最好那次的模型
            if phase == 'valid' and epoch_acc > best_acc:
                # 更新最好的epoch
                best_acc = epoch_acc
                # 更新最好的模型
                best_model_wts = copy.deepcopy(model.state_dict())
                # 统计模型参数信息
                state = {
                  'state_dict': model.state_dict(),#字典里key就是各层的名字，值就是训练好的权重
                  'best_acc': best_acc,
                  'optimizer' : optimizer.state_dict(),#优化器的状态信息
                }
                # 将训练最好的参数信息存储到filename文件中
                torch.save(state, filename)
            if phase == 'valid':
                val_acc_history.append(epoch_acc)
                valid_losses.append(epoch_loss)
                # 学习率衰减（执行7个epoch才会更新学习率）
                scheduler.step(epoch_loss)
            if phase == 'train':
                train_acc_history.append(epoch_acc)
                train_losses.append(epoch_loss)
        
        print('Optimizer learning rate : {:.7f}'.format(optimizer.param_groups[0]['lr']))
        LRs.append(optimizer.param_groups[0]['lr'])
        print()

    # 计算执行完全部epoch(num_epochs)所需时长
    time_elapsed = time.time() - since
    # 打印时间
    print('Training complete in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
    # 打印训练最好的一次的epoch
    print('Best val Acc: {:4f}'.format(best_acc))

    # 训练完后用最好的一次当做模型最终的结果,等着一会测试
    model.load_state_dict(best_model_wts)
    return model, val_acc_history, train_acc_history, valid_losses, train_losses, LRs 
```
