### 图像阈值（二值图）

#### ret, dst = cv2.threshold(src, thresh, maxval, type)

- src： 输入图，只能输入单通道图像，通常来说为灰度图
- dst： 输出图
- thresh： 阈值，不是一个百分比值，就是一个具体的值，一般使用127
- maxval： 当像素值超过了阈值（或者小于阈值，根据type来决定），所赋予的值
- type：二值化操作的类型，包含以下5种类型： cv2.THRESH_BINARY； cv2.THRESH_BINARY_INV； cv2.THRESH_TRUNC； cv2.THRESH_TOZERO；cv2.THRESH_TOZERO_INV

- cv2.THRESH_BINARY           超过阈值部分取maxval（最大值），否则取0
- cv2.THRESH_BINARY_INV    THRESH_BINARY的反转，超过阈值部分取0，否则取maxval(最大值)
- cv2.THRESH_TRUNC            大于阈值部分设为阈值，否则不变
- cv2.THRESH_TOZERO          大于阈值部分不改变，否则设为0
- cv2.THRESH_TOZERO_INV  THRESH_TOZERO的反转，大于阈值的部分设为0，否则不改变



```python
ret, thresh1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

### 图像平滑

```python
img = cv2.imread('lenaNoise.png')

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

在线性滤波（如均值滤波、方框滤波、高斯滤波）中，核心计算过程会涉及到 卷积核权重与图像局部区域的矩阵乘积（点乘），然后将结果累加，再通过归一化或其他方式得到中心像素的新值。

在非线性滤波（如中值滤波）中，不涉及权重乘积，直接通过排序取中值。

注意：不论哪种方式，最终计算完都是替换的图像局部区域中心点的值。


```python
# 均值滤波
# 简单的平均卷积操作，将卷积核和图片矩阵对应位置相乘后相加 / 卷积核矩阵维度（归一化操作）
blur = cv2.blur(img, (3, 3))

cv2.imshow('blur', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 方框滤波
# 基本和均值一样，normalize=True可以选择是否归一化，如果设置为True和均值滤波完全一致，进行归一化操作
box = cv2.boxFilter(img,-1,(3,3), normalize=True)  

cv2.imshow('box', box)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 方框滤波
# 设置为False，那么不进行归一化，如果累加超过255，直接取255作为最终像素值
box = cv2.boxFilter(img,-1,(3,3), normalize=False)  

cv2.imshow('box', box)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 高斯滤波
# 高斯模糊的卷积核里的数值是满足高斯分布，相当于更重视中间的
aussian = cv2.GaussianBlur(img, (5, 5), 1)  

cv2.imshow('aussian', aussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 中值滤波
# 相当于用中值代替
# 这个替换只是替换卷积核中间的那个位置的像素点（一个像素点）替换为 5X5 卷积核范围内像素点从小到大排序中间的那个值
median = cv2.medianBlur(img, 5)

cv2.imshow('median', median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 拼接图片，一起展示
res = np.hstack((blur,aussian,median))
#print (res)
cv2.imshow('median vs average', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 形态学-腐蚀操作

腐蚀主要用于缩小或消除图像中的前景区域边界。

核心思想：用结构元素扫描图像的每个像素，只有当结构元素完全覆盖的区域满足特定条件时，中心像素才会被保留。

二值图像：若结构元素覆盖的所有像素均为前景（例如1或白色），则中心像素保持前景；否则被腐蚀为背景(0或黑色)。

```python
img = cv2.imread('dige.png')

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((3,3),np.uint8)
# 腐蚀函数
erosion = cv2.erode(img,kernel,iterations = 1)

cv2.imshow('erosion', erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pie = cv2.imread('pie.png')

cv2.imshow('pie', pie)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((30,30),np.uint8) 
erosion_1 = cv2.erode(pie,kernel,iterations = 1)
erosion_2 = cv2.erode(pie,kernel,iterations = 2)
erosion_3 = cv2.erode(pie,kernel,iterations = 3)
res = np.hstack((erosion_1,erosion_2,erosion_3))
cv2.imshow('res', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 形态学-膨胀操作

膨胀与腐蚀互为对偶操作，主要用于扩大图像中的前景区域，填充物体内部的孔洞、连接相邻物体或修复断裂的边缘。

核心思想：用结构元素扫描图像的每个像素，只要结构元素覆盖区域与前景有交集，则中心像素会被设为前景。

```python
img = cv2.imread('dige.png')
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((3,3),np.uint8) 
dige_erosion = cv2.erode(img,kernel,iterations = 1)

cv2.imshow('erosion', dige_erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.ones((3,3),np.uint8) 
dige_dilate = cv2.dilate(dige_erosion,kernel,iterations = 1)

cv2.imshow('dilate', dige_dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pie = cv2.imread('pie.png')

kernel = np.ones((30,30),np.uint8) 
dilate_1 = cv2.dilate(pie,kernel,iterations = 1)
dilate_2 = cv2.dilate(pie,kernel,iterations = 2)
dilate_3 = cv2.dilate(pie,kernel,iterations = 3)
res = np.hstack((dilate_1,dilate_2,dilate_3))
cv2.imshow('res', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 开运算与闭运算

核心原理：将腐蚀和膨胀结合到一起，通过 cv2.MORPH_OPEN（先腐蚀再膨胀）/cv2.MORPH_CLOSE（先膨胀再腐蚀） 参数控制

```python
# 开：先腐蚀，再膨胀
img = cv2.imread('dige.png')

kernel = np.ones((5,5),np.uint8) 
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

cv2.imshow('opening', opening)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 闭：先膨胀，再腐蚀
img = cv2.imread('dige.png')

kernel = np.ones((5,5),np.uint8) 
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

cv2.imshow('closing', closing)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
img = cv2.imread('pie.png',cv2.IMREAD_GRAYSCALE)
cv2.imshow("img",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

dst = cv2.Sobel(src, ddepth, dx, dy, ksize)
- ddepth:图像的深度
- dx和dy: 分别表示水平和竖直方向
- ksize: Sobel算子的大小


```python
def cv_show(img,name):
    cv2.imshow(name,img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
# dx=1,dy=0代表目前计算的为水平的（右侧减左侧），不计算竖直的
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)

cv_show(sobelx,'sobelx')
```

以上绘制出来的是一半的圆圈，是因为dx方向一直都是右侧减左侧
白到黑是正数，黑到白就是负数了，所有的负数会被截断成0，所以要取绝对值

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
# 绝对值转换
sobelx = cv2.convertScaleAbs(sobelx)
cv_show(sobelx,'sobelx')
```

```python
# dx=0,dy=1计算竖直方向（下方减上方），不计算水品方向
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
# 绝对值转换
sobely = cv2.convertScaleAbs(sobely)  
cv_show(sobely,'sobely')
```

分别计算x和y，再求和

```python
# 将水平和竖直进行累加计算出最终的结果
sobelxy = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
cv_show(sobelxy,'sobelxy')
```

### Canny边缘检测

Canny边缘检测所需的图像并非二值图，大部分是灰度图像，但是输出是二值化的边缘图（非黑即白，白色表示边缘）。

步骤：
- 1)        使用高斯滤波器，以平滑图像，滤除噪声。

- 2)        计算图像中每个像素点的梯度强度和方向。

- 3)        应用非极大值（Non-Maximum Suppression）抑制，以消除边缘检测带来的杂散响应。

- 4)        应用双阈值（Double-Threshold）检测来确定真实的和潜在的边缘。

- 5)        通过抑制孤立的弱边缘最终完成边缘检测。

#### 高斯金字塔：向下采样方法（缩小）
- 从下往上（向下代表样本数据越来越少）
- 高斯：权重满足高斯函数分布，中间的权重大一些，两边的权重小一些，比较关心中心点的特征信息
- 1/16代表归一化处理，计算完之后需要进行归一化处理，计算出中心点的特征值

![title](Pyramid_2.png)


举例：[4x4] -> [2x2] 原始行中存在 1,2,3,4 ，去除偶数行(2,4)，保留下1行和三行，自然就变为[2x2]

- 每次扩张之后需使用卷积核进行运行，得到最终生成的特征值

### 图像轮廓

输入类型：必须**显式提供二值图像**（例如经过阈值处理和Canny边缘检测后的图像）

原理：cv2.findContours() 的内部实现要求输入是二值图，因为轮廓是寻找白色区域的边界。

图像边缘和轮廓的区别
- 边缘可以是一个点或者是一条线，都可以称之为边缘
- 轮廓必须可以表达出物体的形状特征，所以不能是一个点，必须是以线的形式存在

#### cv2.findContours(img,mode,method)
mode:轮廓检索模式
- RETR_EXTERNAL ：只检索最外面的轮廓；
- RETR_LIST：检索所有的轮廓，并将其保存到一条链表当中；
- RETR_CCOMP：检索所有的轮廓，并将他们组织为两层：顶层是各部分的外部边界，第二层是空洞的边界;
- RETR_TREE：检索所有的轮廓，并重构嵌套轮廓的整个层次;（常用）

method:轮廓逼近方法
- CHAIN_APPROX_NONE：以Freeman链码的方式输出轮廓，所有其他方法输出多边形（顶点的序列）。
- CHAIN_APPROX_SIMPLE:压缩水平的、垂直的和斜的部分，也就是，函数只保留他们的终点部分。

![title](chain.png)

为了更高的准确率，使用二值图像。

```python
# 读取数据
img = cv2.imread('contours.png')
# 获取灰度图
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 获取二值图像，127为一个分界点，超过127获取maxVal也就是255，否则取0
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv_show(thresh,'thresh')
```

```python
# 获取轮廓
# contours：轮廓点集列表。Python列表(list)，每个元素是一个NumPy数组。检测到的所有轮廓的坐标集合，每个轮廓饿坐标点以(x,y)形式的二维数组存储。
# thresh：轮廓层级关系
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

绘制轮廓

### 傅里叶变换的作用

- 高频：变化剧烈的灰度分量，例如边界

- 低频：变化缓慢的灰度分量，例如一片大海

### 滤波

- 低通滤波器：只保留低频，会使得图像模糊。（对应图像中平滑/渐变区域）

- 高通滤波器：只保留高频，会使得图像细节增强。（对应图像中边缘/噪声等细节）





- opencv中主要就是cv2.dft()和cv2.idft()，输入图像需要先转换成np.float32 格式。
- 得到的结果中频率为0的部分会在左上角，通常要转换到中心位置，可以通过shift变换来实现。
- cv2.dft()返回的结果是双通道的（实部，虚部），通常还需要转换成图像格式才能展示（0,255）。

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg',0)

img_float32 = np.float32(img)

dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
# 将低频点转换到中心，越往边上频率越高，频率越高越代表边缘
dft_shift = np.fft.fftshift(dft)
# 得到灰度图能表示的形式
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg',0)

img_float32 = np.float32(img)

dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
# 将低频移动到中心位置
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
# 中心位置
crow, ccol = int(rows/2) , int(cols/2)    

# 低通滤波
# 使用0来构建一个和原图尺寸大小一致的严模mask（抑制所有频率）
mask = np.zeros((rows, cols, 2), np.uint8)
# 保留中心低频
mask[crow-30:crow+30, ccol-30:ccol+30] = 1

# IDFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Result'), plt.xticks([]), plt.yticks([])

plt.show()                
```

```python
img = cv2.imread('lena.jpg',0)

img_float32 = np.float32(img)

dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
# 中心位置
crow, ccol = int(rows/2) , int(cols/2)     

# 高通滤波
# 初始化为1（保留所有频率）
mask = np.ones((rows, cols, 2), np.uint8)
# 将中心区域设置为0（抑制低频）
mask[crow-30:crow+30, ccol-30:ccol+30] = 0

# IDFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Result'), plt.xticks([]), plt.yticks([])

plt.show()    
```
